```python
#question 1
class Monster:
  monsterCount=0
  def __init__(self,name,power):
    self.name=name
    self.power=power
    Monster.monsterCount+=1
    self.alive=True
  def get_details(self):
    return(f'Name:{self.name}\nPower:{self.power}\nAlive:{self.alive}')
  
  def attack(self,*arg):
    if self.alive==True:
      if len(arg)==0:
        print('No monsters to attack')
      elif len(arg)==1:
        b=arg[0]
        if self.power>b.power:
          print(f'Attack successful.{self.name} defeated {b.name}')
          Monster.monsterCount-=1
          b.alive=False
      elif len(arg)==2:
        b=arg[0]
        c=arg[1]
        if b.alive==False and c.alive==True:
          if c.power>self.power:
            print (f'Cannot attack {b.name}. Its not alive.\nAttack unsuccessful.{self.name} was\ndefeated by {c.name}')
          Monster.monsterCount-=1
          self.alive=False 
               
    else:
      print(f'{self.name} is not alive to attack.')
monster1 = Monster('Godzilla', 40)
monster2 = Monster('Hydra', 30)
monster3 = Monster('KingKong', 50)
print(f"Number of monstersalive:{Monster.monsterCount}")
print('1--------------------------------')
print(monster1.get_details())
print('2--------------------------------')
print(monster2.get_details())
print('3--------------------------------')
print(monster3.get_details())
print('4--------------------------------')
monster1.attack()
print('5--------------------------------')
monster1.attack(monster2)
print('6--------------------------------')
monster1.attack(monster2, monster3)
print('7--------------------------------')
print(f"Number of monstersalive:{Monster.monsterCount}")
print('8--------------------------------')
print(monster2.get_details())
print('9--------------------------------')
monster2.attack(monster1)
```
