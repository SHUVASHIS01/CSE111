```python
#task1
class Student:
  ID=0
  def __init__ (self,name,dept,age,cg):
    self.name=name
    self.dept=dept
    self.age=age
    self.cg=cg
    Student.ID+=1
  @classmethod
  def from_String(cls, info):
    name, dept, age, cg = info.split('-')
    return cls(name,dept,int(age),float(cg))
  def showDetails(self):
    print(f'ID: {Student.ID}\nName: {self.name}\nDepartment: {self.dept}\nAge: {self.age}\nCGPA: {self.cg} ')


s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails() 
print('Class variables are shared across all objects while instance variables are for data unique to each instance.')
print('Instance methods need a class instance and can access the instance through self but Class methods do not need a class instance. ')
```

```python
#task2
class Assassin:
  suc_rate=100
  c=0
  def __init__(self,name,srate):
    self.name=name
    self.srate=srate
    Assassin.c+=1
  @classmethod
  def failureRate(cls,name,rate_fall):
    return cls(name, cls.suc_rate-rate_fall)
  @classmethod
  def failurePercentage(cls,name,rate_fall):
    a=int(rate_fall[:-1:])
    return cls(name,cls.suc_rate-a)
  def printDetails(self):
    print(f'Name: {self.name}\nSuccess rate: {self.srate}%\nTotal number of Assassin: {Assassin.c} ')




john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#task5
class Employee:
  crnt_year=2022
  def __init__ (self,name,workingPeriod):
    self.name=name
    self.workingPeriod=workingPeriod
  @classmethod
  def employeeByJoiningYear(cls,name,year):
    return cls(name,Employee.crnt_year-year)
  @staticmethod
  def experienceCheck(exp,gender):
    if exp<3 and gender=='male':
      return 'He is not experienced'
    elif exp<3 and gender=='female':
      return 'She is not experinced'
    elif exp>=3 and gender=='male':
      return 'He is experienced'
    elif exp>=3 and gender=='female':
      return 'She is experienced'



employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#task7
class Cat:
  Number_of_cats=0
  def __init__ (self,clr,wrk):
    self.clr=clr
    self.wrk=wrk
    Cat.Number_of_cats+=1
  @classmethod
  def no_parameter(cls):
    return cls('White', 'sitting')
  @classmethod
  def first_parameter(cls,clr):
    return cls(clr,'sitting')
  @classmethod
  def second_parameter(cls,wrk):
    return cls('Grey',wrk)
  def changeColor(self,clr):
    self.clr=clr
  def printCat(self):
    print(f'{self.clr} cat is {self.wrk}')
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task8
import math
class Cylinder:
  radius, height = 5, 18
  def __init__(self,radius,height):
    print(f'Default radius={Cylinder.radius} and height={Cylinder.height} ')
    self.radius=radius
    self.height=height    
    print(f'Updated: radius={self.radius} and height={self.height}')
  @staticmethod
  def area(rad,h):
    print(f'Area: {(2*math.pi*rad*rad)+(2*math.pi*rad*h)}')
  @staticmethod
  def volume(rad,h):
    print(f'Volume: {math.pi*rad*rad*h}')
  @classmethod
  def swap (cls,rad,h):
    a,b=h,rad
    radius,height=b,a
    return cls(b,a)
  @classmethod
  def changeFormat(cls,inf):
    r,h=inf.split('-')
    radius,height=r,h
    return cls(float(r),float(h))


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#task10
class SultansDine:
  total_brancs,total_sell,l=0,0,[]
  def __init__ (self,loc):
    self.loc=loc
    SultansDine.total_brancs+=1
  def sellQuantity(self,n):
    if n<10:
      self.Branch_sell=n*300
    elif n<20:
      self.Branch_sell=n*350
    else:
      self.Branch_sell=n*400
    SultansDine.total_sell+=self.Branch_sell
  def branchInformation(self):
    print(f'Branch Name: {self.loc}\nBranch Sell: {self.Branch_sell}  ')
    SultansDine.l.append(self.loc)
    SultansDine.l.append(self.Branch_sell)
  @classmethod
  def details (cls):
    print(f'Total Number of branch(s): {cls.total_brancs} ')
    print(f'Total Sell: {cls.total_sell} ')
    for i in range (0,len(SultansDine.l),2):
        print(f'Branch Name: {cls.l[i]}, Branch sell {cls.l[i+1]} ')
        print(f'Branch consists of total sells: {(cls.l[i+1]/cls.total_sell)*100:.2f}% ')




SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
