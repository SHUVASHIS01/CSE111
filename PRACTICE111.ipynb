```python
def addition_tup(lst,num):
  l=[]
  k=[]
  for i in lst:
    k.append(i)
    k.append(i+num)
    l.append(tuple(k))
  print(tuple(l))
addition_tup([-1,-2, -3, 0, 5], -5.2)
```

```python
def common_value_in_tups(t1,t2):
  l=list(t1)
  k=list(t2)
  c=0
  for i in l:
    if i in k:
      c+=1
  if c>0:
    return True
  else:
    return False
print(common_value_in_tups((-1,0,2),(5,10.2,0)))
```

```python
def swap_keys_values(n):
  d={}
  for k,v in n.items():
    if v not in d:
      d[v]=k
  return d
swap_keys_values({'a':1, 'b': 2, 'c':7})
```

```python
def greater_than_five(n):
  d={}
  l=[]
  for k,v in n.items():
    if v>5:
      l+=k
  return l
greater_than_five({'a':1, 'b': 2, 'c':7})
```

```python
class Hotel:
  def __init__(self, hname):
    self.hname=hname
    self.scount=0
    self.gcount=0
    self.d={}
    self.k={}
  def addStuff(self,name,age,phone='000'):
    self.name=name
    self.age=age
    self.phone=phone
    self.scount+=1
    print(f'Staff With ID {self.scount} is added')
  def getStuffById(self,id):
    self.id=id
    self.d[self.id]=[self.name, self.age, self.phone]
    return f'Staff ID: {self.id}\nName: {self.name}\nAge: {self.age}\nPhone no.:{self.phone}'
  def addGuest(self,name,age,phone='000'):
    self.name=name
    self.age=age
    self.phone=phone
    self.gcount+=1
    print(f'Staff With ID {self.scount} is added')
  def getGuestById(self,id):
    self.id=id
    self.k[self.id]=[self.name, self.age, self.phone]
    return f'Staff ID: {self.id}\nName: {self.name}\nAge: {self.age}\nPhone no.:{self.phone}'
  def allStaffs(self):
    print(f'All staffs:\nNumber of staff: {self.scount}  ')
    for k,v in self.d.items():
      print(f'Staff id: {k} Name: {v[0]} Age: {v[1]} Phone no.: {v[2]} ')
  def allGuest(self):
    print(f'All staffs:\nNumber of staff: {self.gcount}  ')
    for k,v in self.k.items():
      print(f'Staff id: {k} Name: {v[0]} Age: {v[1]} Phone no.: {v[2]} ')
h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest('Carol',35,'123')
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, '431')
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```
