```python
#TASK1
l=[]
i=0
while i>=0:
  n=input()
  l.append(n)
  if n=="STOP":
    break
  i+=1
k=[]
for i in l:
  if i not in k:
    k.append(i)
for i in k:
  if i!='STOP':
    v=l.count(i)
    print(i,'-',v,'times')
```

```python
#TASK3
i=0
while i>=0:
  N=input()
  if N=='STOP':
    break
  else:
    n=N.split(' ')
    l=[]
    for j in range (0,len(n)-1):
      d=int(n[j])-int(n[j+1])
      d=abs(d)
      l.append(d)
    if all(x <= (len(n)-1) for x in l):
      print('UB Jumper')
    else:
      print('Not UB Jumper')
  i+=1
```

# **Dictionary & Tuple**

```python
#TASK1
d1={}
d2={}
n=input()
m=input()
s=n.split(', ')
p=m.split(', ')
for i in s:
  i=i.split(':')
  d1[i[0]]=int(i[1])
for j in p:
  j=j.split(':')
  d2[j[0]]=int(j[1])
for k in d2:
  if k in d1:
    d1[k]=d1[k]+d2[k]
  else:
    d1[k]=d2[k]
print(d1)
l=[]
k=[]
for v in d1.values():
  l.append(v)
for x in l:
  if x not in k:
    k.append(x)
print('Values:',tuple(sorted(k)))
```

# **Function**

```python
def BMI(a,b):
  a=a/100
  cal_bmi=b/(a*a)
  if cal_bmi<18.5:
    return 'Score is {:.1f}. You are Underweight'.format(cal_bmi)
  elif  18.5<=cal_bmi<=24.9:
    return 'Score is {:.1f}. You are Normal'.format(cal_bmi)
  elif  25<=cal_bmi<=30:
    return 'Score is {:.1f}. You are Overweight'.format(cal_bmi)
  elif cal_bmi>30:
    return 'Score is {:.1f}. You are Obese'.format(cal_bmi)
print(BMI(int(input()),int(input())))
```

```python
#TASK2
def fname(minimum,maximum,divisor):
  s=0
  for i in range (minimum,maximum):
    if i%divisor==0:
      s+=i
  return s
print(fname(int(input()),int(input()),int(input())))
```

```python
def common_value_in_tups(t1,t2):
  l=list(t1)
  k=list(t2)
  c=0
  for i in l:
    if i in k:
      c+=1
  if c>0:
    return True
  else:
    return False
print(common_value_in_tups((-1,0,2),(5,10.2,0)))
```
