```python
class Chitchat:
  def __init__(self,arg,country='Bangladesh'):
    m=arg.split(',')
    self.name=m[0]
    self.id=m[1]
    self.age=m[2]
    self.gender=m[3]
    self.country=country
    self.d={}
    self.s=[self.name+'_'+str(self.id)]
    self.frnd=0
    
    if self.country=='Bangladesh':
      print('Your request is being processed...Account creation successful. But you have no friends. Add some friends and chill.')
    else:
      print('Sorry! Account creation is not allowed outside Bangladesh.')
  def AddFriend(self,fname,relation=None):
    self.fname=fname
    self.relation=relation      
    self.d[self.relation]=self.s
    if self.relation==None:
      return 'Request Rejected! Relationship required.'
    else:
      return 'Friend request sent. Pending for approval.'
  def sendMessage(self,obj,msg):
    self.obj=obj
    self.msg=msg
    if self.frnd==0:
      print(f'Ooopps! You cannot chat with this person (id -> {self.obj.id}).')
    else:
      print(f'Sent messege: {self.msg}\nReciever Details==> Name:{self.name} ID:{self.id} Age: {self.age} gender: {self.gender} ')
  def approveFriendReq(self):
    self.frnd+=1
    print(f'Congratulations! {self.name} and {self.name} can chat now.')
  def changeStatus(self):
    print('Account activated/deactivated.')
  def showNetwork(self):
    print(f'{self.name}s network: {self.d} ')
  









print("Welcome to Chitchat! Create an account, have fun!")
Wahid = Chitchat("Wahid,00123,23,Male")
print("1======================================")
Maliha = Chitchat("Maliha,00124,25,Female", "Malaysia")
print("2======================================")
Sadik = Chitchat("Sadik,00125,26,Male", "Bangladesh")
print("3======================================")
print(Wahid.AddFriend(Sadik, "Brother"))
print("4======================================")
Wahid.sendMessage(Sadik, "Hi, bro. How are you?")
print("5======================================")
Sadik.approveFriendReq()
print("6======================================")
Wahid.sendMessage(Sadik, "Hi, bro. How are you?")
print("7======================================")
Wahid.changeStatus()
print("8======================================")
Sadik.sendMessage(Wahid, "Hello, I'm fine.")
print("9======================================")
Sadik.showNetwork()
print("10=====================================")
Lamia = Chitchat("Lamia,00126,26,Female")
print("11=====================================")
print(Sadik.AddFriend(Lamia))
print("12=====================================")
```
